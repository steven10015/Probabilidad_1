## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np

num_ensayos = 1000

# Hacemos la simulación de los lanzamientos para cada dado
dado_1 = np.random.choice([1, 2, 3, 4, 5, 6], size = num_ensayos)
dado_2 = np.random.choice([1, 2, 3, 4, 5, 6], size = num_ensayos)

# Almacenamos todas las sumas en una nueva variable
suma_de_dados = (dado_1 + dado_2)

# Variable vacia para ser contador
conteo_de_suma = 0
# Bucle para filtrar y almacenar nuestros resultados de interés
for suma in suma_de_dados: 
    if suma > 7 or suma %2 == 0:
        conteo_de_suma += 1

# Visualizamos resultados
print(f"La cantidad de veces que la suma fue mayor a 7 o un número par fueron {conteo_de_suma} veces.")

probabilidad = ((conteo_de_suma))/num_ensayos
print(f"La probabilidad es por tanto un {probabilidad}.")

porcentaje = probabilidad * 100
print(f"El procentaje de veces que la suma de ambos dados fue mayor a 7 o un número par es de un {round(porcentaje, 2)}%.")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
# Crear la caja de bolas
# Diccionario vacio para almacenar
caja_de_bolas = {}
# Bucleo donde por medio de condiciones dejamos seteadas la cantidad de bolas de determinado color
for i in range(60):
    # Recordatorio, escribir caja_de_bolas[i], hace que se guarde en el diccionario como un valor
    if i < 10:
        caja_de_bolas[i] = "Blancas"
    elif i < 30:
        caja_de_bolas[i] = "Rojas"
    else:
        caja_de_bolas[i] = "Verdes"

# Se define una funcion para probar el experimento un determinado número de veces
# Se dejan escritos ambos conteos
def escoger_bolas(numero_simulaciones):
    conteo_1 = 0
    conteo_2 = 0

# Un loop que itere por la cantidad de ensayos
    for _ in range(numero_simulaciones):
        # Extraer 5 bolas sin reemplazo
        indices = np.random.choice(range(60), size=5, replace=False)

        # Lista por comprensión
        # ball_box[i] son los colores en el diccionario
        # para cada valor en indices, se toma su valor correspondiente en caja de bolas
        colores = np.array([caja_de_bolas[i] for i in indices])
        bolas_blancas = sum(colores == "Blancas")
        bolas_rojas = sum(colores == "Rojas")
        bolas_verdes = sum(colores == "Verdes")

        # Primera posibilidad, 3 bolas blancas y 2 rojas
        if (bolas_blancas == 3) and (bolas_rojas == 2):
            conteo_1 += 1
        # Segunda probabilidad, todas del mismo color
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            conteo_2 += 1
    # Se almacenan dos valores finales
    return conteo_1 / numero_simulaciones, conteo_2 / numero_simulaciones

probabilidad = escoger_bolas(10000)

print(f"La probabilidad de tomar 3 bolas blancas y 2 rojas es de: {np.round(probabilidad[0] * 100, 2)}%")
print(f"La probabilidad de que todas las bolas escogidas sean del mismo color es de: {np.round(probabilidad[1] * 100, 2)}%")
```
